```python
def plot_hist(sample_size):
    z=[]
    for i in range(8000):
        r = dweibull.rvs(c=2, size=sample_size)
        mean_value = statistics.mean(r)
        z.append(mean_value)
    print("Standard Deviation: ",np.std(z,axis=0))
    plt.xlim(-2,2)
    plt.ylim(0,5)
    plt.hist(z,bins=2000,density=True)
    plt.title("Sample size is %s" %sample_size)
```

#### We can see that as the sample size increases, the standard deviation decreases.

Now, plotting the standard deviation as a function of the sample size.

```python
def std_of_sample(sample_size):
    z=[]
    for i in range(500):
        r = dweibull.rvs(c=2, size=sample_size)
        mean_value = statistics.mean(r)# It gives the sample mean
        z.append(mean_value)
    std_value = np.std(z,axis=0)
    return std_value
```

```python
X = np.arange(start=1,stop=100,step=1)
print(X)
for x in X :
    plt.scatter(x,std_of_sample(x))
plt.xlabel("Sample Size")
plt.ylabel("Standard Deviation")
plt.title("Comparing std V/s Sample size plot with y  = 1/sqrt(x)")

r = np.linspace(0,100,1000)
plt.plot(r,np.power(r,-0.5))
```

 **The std. deviation reduces as the square root of the sample size**

```python
def val(sample_size,c_val):
    z=[] # z here is the array storing the sample means
    for i in range(500):
        r = dweibull.rvs(c=c_val, size=sample_size)
        mean_value = statistics.mean(r)
        val = np.sqrt(sample_size)*mean_value
        z.append(val)
    return z
```

```python
def mean_value(D):
    N = len(D)
    #print(N)
    sum = 0
    for x in D:
        sum = sum + x
        #print(x)
    mean_val = sum/N
    return mean_val    
```

```python
def std_value(D):
    mean_val = mean_value(D)
    N = len(D)
    std_val = 0
    for x in D:
        std_val = std_val + ((x-mean_val)**2)/(N-1)
    std_val = np.sqrt(std_val)
    return std_val
```

```python
def plot_hist(arr):
    plt.hist(arr, bins = 20, density = True)
```

```python
def plot_pdf(x):
    mu = 0
    sigma = 1
    #print(np.sqrt(n))
    plt.plot(x, stats.norm.pdf(x, mu, sigma),label='Normal distribution')
    plt.legend(loc='best')
```

```python
def plot_t(x,D,n):
    df = n-1
    mu = 0
    sigma = 1
    plt.plot(x,dist_t.pdf(x,df,mu,sigma),label='t-distribution')
    plt.legend(loc='best')
    plt.show()
```

```python
def plots(sample_size,c_val):
    D = val(sample_size,c_val)
    plot_hist(D)
    plt.xlabel("Sample mean")
    plt.ylabel("Frequency of sample means")
    x = np.linspace(-3,3,500)
    plot_pdf(x)
    plot_t(x,D,sample_size)
```
