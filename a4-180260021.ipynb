```python
# This function will return either 1 or 0 indicating True or False i.e. whether O + D = I or not.
def rolling_event():
    
    Octahedron = np.arange(1,8)
    Dodecahedron = np.arange(1,12)
    Icosahedron = np.arange(1,20)
    
    O = random.choice(Octahedron)
    D = random.choice(Dodecahedron)
    I = random.choice(Icosahedron)
    #print("O+D ",O+D)
    #print("I ",I)
    if(O + D == I):
        return 1
    else:
        return 0
```

```python
# This function computes the probability of O + D = I by repeating the rolling_event "n" times
# "n" is the no. of repetitions of the rolling event
# Our outcomes here imply true or false in the form of 1 or 0
def experiment(n):
    positive_outcomes = 0
    
    for i in range(n):
        if(rolling_event()==1):
            positive_outcomes += 1
    
    #positive_outcomes = np.sum(outcomes)
    probability = positive_outcomes/n
    #print("probability ",probability)
    
    return probability
```

# B. Analysing the data sets derived

```python
def mean_value(D):
    N = len(D)
    #print(N)
    sum = 0
    for x in D:
        sum = sum + x
        #print(x)
    mean_val = sum/N
    return mean_val    
```

```python
def std_value(D):
    mean_val = mean_value(D)
    N = len(D)
    std_val = 0
    for x in D:
        std_val = std_val + ((x-mean_val)**2)/(N-1)
    std_val = np.sqrt(std_val)
    return std_val
```

```python
def plot_boxplot(D):
    plt.boxplot(D)
 #   plt.ylim(-0.05,0.15)
    plt.grid()
    plt.xlabel('Data')
    plt.ylabel('Data values')
```

```python
def sample(n,D):
    sample=[]
    N = len(D)
    indices = np.random.randint(0,N,n)
    for index in indices:
        sample.append(D[index])
    return sample
```

```python
def sample_mean_array(M, n, D):
    sample_mean_arr = []
    for i in range(M):
        sample_mean_arr.append(mean_value(sample(n,D)))
    return sample_mean_arr
```

```python
def sample_std_array(M, n, D):
    sample_std_arr = []
    for i in range(M):
        sample_std_arr.append(std_value(sample(n,D)))
    return sample_std_arr
```

```python
def plot_hist(arr):
    plt.hist(arr, bins = 15,density = True)
```

```python
def plot_pdf(x,D,n):
    mu = mean_value(D)
    sigma = std_value(D)/np.sqrt(n)
    #print(np.sqrt(n))
    plt.plot(x, stats.norm.pdf(x, mu, sigma),label='Normal distribution')
    plt.legend(loc='best')
```

```python
def plot_t(x,D,n):
    df = n-1
    mu = mean_value(D)
    sigma = std_value(D)/np.sqrt(n)
    plt.plot(x,dist_t.pdf(x,df,mu,sigma),label='t-distribution')
    plt.legend(loc='best')
    plt.show()
```

#### Calculating Confidence Intervals

https://www.investopedia.com/terms/t/tdistribution.asp

```python
# This function returns t-score i.e. t(n-1,alpha/2)*s/sqrt(n)
def calc_t_dist_factor(sample,ci_val):
    
    alpha = 1.0 - ci_val
    ppf_arg = 1.0 - 0.5*alpha
    n = len(sample)
    
    # t-distribution with (n - 1) degrees of freedom
    td = dist_t(n - 1)
    ci = td.ppf(ppf_arg)*std_value(sample)/np.sqrt(n)
    
    return ci
```

```python
# This function returns z-score i.e. z(alpha/2)*s/sqrt(n)
def calc_n_dist_factor(sample,ci_val):
    
    alpha = 1.0 - ci_val
    ppf_arg = 1.0 - 0.5*alpha
    n = len(sample)
    
    ci = dist_norm.ppf(ppf_arg)*std_value(sample)/np.sqrt(n)
    
    return ci
```

```python
# Now calculating the Confidence Intervals for sample_size = n, value_of_confidence_interval = ci_val,
# M =100(no. of repetitions of experiment)

def print_ci(n, ci_val, M, D):
    print("n = ",n)
    lies_in_t = 0
    lies_in_n = 0
    for i in range(M):
        sample1 = sample(n,D)
        ci_t = calc_t_dist_factor(sample1,ci_val)
        ci_n = calc_n_dist_factor(sample1,ci_val)
        mean_sample = mean_value(sample1)
        mean_data = mean_value(D)
        print("\n")
        print(i,".")
        print("Approximate {}% confidence interval of the mean for T-Distribution: {:.4f} +- {:.4f}".format(ci_val*100, mean_sample, ci_t))
        print("[",mean_sample-ci_t,", ",mean_sample+ci_t,"]")
        if mean_data >= (mean_sample - ci_t) and mean_data <= (mean_sample + ci_t):
            lies_in_t = lies_in_t + 1
            print("True mean ({:.4f}) lies in the confidence {}% interval for T-Distribution.".format(mean_data, 100*ci_val))
        else:
            print("True mean ({:.4f}) does not in the confidence {}% interval for T-Distribution!".format(mean_data, 100*ci_val))
            
        print("Approximate {}% confidence interval of the mean for Normal-Distribution: {:.4f} +- {:.4f}".format(ci_val*100, mean_sample, ci_n))
        print("[",mean_sample-ci_n,", ",mean_sample+ci_n,"]")
        if mean_data >= (mean_sample - ci_n) and mean_data <= (mean_sample + ci_n):
            lies_in_n = lies_in_n + 1
            print("True mean ({:.4f}) lies in the confidence {}% interval for Normal-Distribution.".format(mean_data, 100*ci_val))
        else:
            print("True mean ({:.4f}) does not in the confidence {}% interval for Normal-Distribution!".format(mean_data, 100*ci_val))
        
    percent_t_d = [(lies_in_t/M)*100, (lies_in_n/M)*100]
    return percent_t_d
```
