```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

For an exponentially distributed random variable with lambda = 1 :<br>
 f(x) = exp(-x) , x>0 and 0,otherwise

```python
def pdf_for_exp_dist(x):
    y_values = []
    for x_val in x:
        if x_val>0:
            y_values.append(np.exp(-1*x_val))
        else:
            y_values.append(0)
        
    return y_values
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.exp(-x))
```

```python
x = np.linspace(-10,10,2000)
plt.plot(x,pdf_for_exp_dist(x))
plt.xlabel('x')
plt.ylabel('f(x) i.e. PDF(x)')
```
